## Bias in Neural Networks

A neuron computes:

$$
z = w_1x_1 + w_2x_2 + \cdots + w_nx_n + b
$$

where:

- $w$ = weights  
- $x$ = inputs  
- $b$ = bias

```python
# Example:
# y_hat = 2000 * years_of_experience + 30000

years_of_experience = 0

y_hat = 2000 * years_of_experience + 30000

print("Predicted Salary:", y_hat)
```

## Weighted Sum

The neuron first computes:

$$
z = W \times X + b
$$

This value is called the pre-activation output.

```python
# Inputs
x1 = 85
x2 = 3
x3 = 35

# Weights
w1 = -0.2
w2 = 4
w3 = -0.1

# Bias
b = 5

# Weighted sum
z = (w1 * x1) + (w2 * x2) + (w3 * x3) + b

print("Pre-activation output z =", z)
```


## ReLU Activation Function

$$
\mathrm{ReLU}(z) = \max(0, z)
$$

If the input is negative, the output becomes 0.

If the input is positive, it passes unchanged.

```python
def relu(z):
    return max(0, z)

z = -3.5

a = relu(z)

print("Activation Output:", a)
```

```python
# Inputs
weight = 85
activity = 3
age = 35

# Weights
w1 = -0.2
w2 = 4
w3 = -0.1

# Bias
b = 5

# Weighted sum
z = (w1 * weight) + (w2 * activity) + (w3 * age) + b

# ReLU activation
a = max(0, z)

print("z =", z)
print("ReLU Output =", a)
```

```python
# Increased activity level

weight = 85
activity = 4
age = 35

z = (-0.2 * weight) + (4 * activity) + (-0.1 * age) + 5

a = max(0, z)

print("z =", z)
print("ReLU Output =", a)
```

## Sigmoid Function

$$
f(x) = \frac{1}{1 + e^{-x}}
$$

The output is always between 0 and 1.

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

print("Sigmoid(2) =", sigmoid(2))
print("Sigmoid(-2) =", sigmoid(-2))
```

## Xavier Initialization

$$
\mathrm{Var}(w) = \frac{2}{l_{in} + l_{out}}
$$

```python
l_in = 2
l_out = 2

variance = 2 / (l_in + l_out)

print("Variance =", variance)
```
