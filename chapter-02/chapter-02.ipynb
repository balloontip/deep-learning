```python
import math

# The sigmoid function: σ(x) = 1 / (1 + e^(-x))
def sigmoid(x):
    return 1 / (1 + math.exp(-x))

# Example 1: x = 150
x1 = 150
result1 = sigmoid(x1)
print(f"sigmoid({x1}) = {result1}")
print(f"  Since e^(-150) is extremely close to zero, the result is almost exactly 1.\n")

# Example 2: x = -2
x2 = -2
result2 = sigmoid(x2)
print(f"sigmoid({x2}) = {result2:.4f}")
```

```python
import math

# Binary Cross-Entropy Loss:
# Loss = -[y * log(y_hat) + (1 - y) * log(1 - y_hat)]

def cross_entropy_loss(y, y_hat):
    return -(y * math.log(y_hat) + (1 - y) * math.log(1 - y_hat))

print("Binary Cross-Entropy Loss Formula:")
print("Loss = -[y * log(ŷ) + (1 - y) * log(1 - ŷ)]")
print("\ny = true label (0 or 1)")
print("ŷ = model's predicted probability (between 0 and 1)")
print("log = natural logarithm (base e)")
```

```python
# Scenario 1: True Label y = 1

# Model predicts 0.95 (Confident and Correct)
loss1 = cross_entropy_loss(1, 0.95)
print(f"y=1, ŷ=0.95 -> Loss = {loss1:.4f}  (Low - confident and correct)")

# Model predicts 0.60 (Less Confident, Correct Direction)
loss2 = cross_entropy_loss(1, 0.60)
print(f"y=1, ŷ=0.60 -> Loss = {loss2:.4f}  (Higher - less confident)")

# Model predicts 0.10 (Confident and Incorrect)
loss3 = cross_entropy_loss(1, 0.10)
print(f"y=1, ŷ=0.10 -> Loss = {loss3:.4f}  (Huge - confident and wrong)")
```

```python
# Scenario 2: True Label y = 0

# Model predicts 0.05 (Confident and Correct)
loss4 = cross_entropy_loss(0, 0.05)
print(f"y=0, ŷ=0.05 -> Loss = {loss4:.4f}  (Low - confident and correct)")

# Model predicts 0.85 (Confident and Incorrect)
loss5 = cross_entropy_loss(0, 0.85)
print(f"y=0, ŷ=0.85 -> Loss = {loss5:.4f}  (High - confident and wrong)")

print("\nCross-entropy cares HOW SURE you were about a wrong answer.")
print("Confidence matters: be confident only when truly sure, cautious when not.")
```

```python
# Mean Squared Error (MSE)
# Formula: MSE = (1/n) * Σ(y - ŷ)²

# Example 1: One prediction
actual1 = 30
predicted1 = 25
mse1 = (actual1 - predicted1) ** 2 / 1
print(f"Example 1: Actual={actual1}, Predicted={predicted1}")
print(f"  MSE = ({actual1} - {predicted1})² / 1 = {mse1:.2f}\n")

# Example 2: Two predictions
actuals = [30, 50]
predictions = [25, 55]
n = len(actuals)
mse2 = sum((a - p) ** 2 for a, p in zip(actuals, predictions)) / n
print(f"Example 2: Actuals={actuals}, Predictions={predictions}")
print(f"  MSE = [({actuals[0]}-{predictions[0]})² + ({actuals[1]}-{predictions[1]})²] / {n} = {mse2:.2f}")
print("\nThe lower the MSE, the better our predictions are.")
```

```python
# Cross-Entropy Loss examples from Math Box
# Classifying "Fit" (1) vs "Unfit" (0)

# Example 1: Correct class is 1, model predicts 0.9 (pretty confident)
loss_ex1 = cross_entropy_loss(1, 0.9)
print(f"Example 1: y=1, ŷ=0.9 -> Loss = {loss_ex1:.4f}")

# Example 2: Correct class is 1, model predicts 0.1 (completely wrong and confident)
loss_ex2 = cross_entropy_loss(1, 0.1)
print(f"Example 2: y=1, ŷ=0.1 -> Loss = {loss_ex2:.4f}")

print(f"\nIf you're wrong but confident, the model gets scolded much harder.")
print(f"That's Cross-Entropy Loss doing its job.")
```
