```python
# Traditional House Price Formula from Chapter 1
# price = (size × 100) + (num_bedrooms × 20,000) + 60,000

size = 2000          # square feet
num_bedrooms = 3
bias = 60000

price = (size * 100) + (num_bedrooms * 20000) + bias
print(f"Estimated House Price: ${price}")
```

```python
# Electricity Bill Formula from Chapter 1
# bill = (kWh × 0.12) + (num_appliances × 5) + bias

kWh = 500
num_appliances = 10
bias = 20

bill = (kWh * 0.12) + (num_appliances * 5) + bias
print(f"Estimated Electricity Bill: ${bill}")
```

```python
# Supervised Learning Example: Sentiment Labels (from Chapter 1)

reviews = [
    {"Review Text": "The product is amazing!", "Sentiment Label": "Positive"},
    {"Review Text": "It broke after one day of use.", "Sentiment Label": "Negative"},
    {"Review Text": "Customer service was very helpful.", "Sentiment Label": "Positive"},
]

print(f"{'Review Text':<45} {'Sentiment Label'}")
print("-" * 60)
for r in reviews:
    print(f"{r['Review Text']:<45} {r['Sentiment Label']}")
```

```python
# Traditional Method: The BMI Formula
# BMI = weight (kg) / height² (m²)

weight = 85   # kg
height = 1.80 # meters

bmi = weight / (height ** 2)
print(f"BMI = {weight} / {height}² = {bmi:.1f}")
```

```python
# Alex vs Mark: Same BMI, Different Health Profiles

data = [
    ["Alex", 1.80, 85, 26.2, "10%", "45%", "High (Athlete)"],
    ["Mark", 1.80, 85, 26.2, "25%", "30%", "Low (Sedentary)"],
]

header = f"{'Name':<8} {'Height':<10} {'Weight':<10} {'BMI':<8} {'Body Fat':<10} {'Muscle':<10} {'Activity'}"
print(header)
print("-" * 75)
for row in data:
    print(f"{row[0]:<8} {row[1]:<10} {row[2]:<10} {row[3]:<8} {row[4]:<10} {row[5]:<10} {row[6]}")

print("\nSame BMI (26.2) but very different health profiles!")
```

```python
# Real Estate Example: Two Identical-Size Houses, Very Different Prices

houses = [
    ["House A", 2000, "9/10", "Low", "Newly Renovated", "High", "$500,000"],
    ["House B", 2000, "5/10", "High", "Needs Work", "Low", "$320,000"],
]

header = f"{'House':<10} {'Sq Ft':<8} {'School':<8} {'Crime':<8} {'Renovations':<18} {'Demand':<8} {'Price'}"
print(header)
print("-" * 80)
for h in houses:
    print(f"{h[0]:<10} {h[1]:<8} {h[2]:<8} {h[3]:<8} {h[4]:<18} {h[5]:<8} {h[6]}")

print("\nSimple formula: Price = Sq Ft × Price per Sq Ft")
print("But this ignores location, crime, schools, renovations, and market trends!")
```

## Chapter 1 Key Takeaways
- Traditional formulas (price, BMI, bills) are simple but limited
- Deep learning models learn patterns from real data
- Same inputs (like BMI) can mean very different things — AI considers more features
- The pseudocode pattern is always: Load Data → Train Model → Save → Load → Predict
- Labeled data (supervised learning) is how models learn: inputs paired with correct answers
